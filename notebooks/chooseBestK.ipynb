```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px

# Load PCA-transformed data
df_pca = pd.read_csv('data/pca.csv')
```

```python
# Choosing best number of clusters
from sklearn.metrics import calinski_harabasz_score

k_range = range(10, 50)
ch_scores = []
for k in k_range:
    
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(df_pca)
    
    # Calculate CH INDEX
    score = calinski_harabasz_score(df_pca, labels)
    ch_scores.append(score)
    
    print(f"k={k}, Score={score:.2f}")
```

```python
# Plot CH INDEX scores
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(k_range), y=ch_scores, mode='lines+markers', name='CH Index'))
fig.update_layout(title='Calinski-Harabasz Index vs Number of Clusters',
                    xaxis_title='Number of Clusters (k)',
                    yaxis_title='Calinski-Harabasz Index')
fig.show()
```
