```python
# Loading the dataset

df_original = pd.read_csv('data/zfish.csv')
```

```python
# Info
df_original.info()
```

```python
# Head
df_original.head(3)
```

```python
# How many unique types are in the dataset?
print("different types:", sorted(df_original['Type'].unique().tolist()))
print("number of different types:", df_original['Type'].nunique())
```

```python
# How many unique RGBs are in the dataset?

for r,g,b in zip(df_original['R'].unique(), df_original['G'].unique(), df_original['B'].unique()):
    print(f"R: {r}, G: {g}, B: {b}") 
```

```python
# Which types in which RGBs combination?
type_rgb_combinations = df_original.groupby(['Type', 'R', 'G', 'B']).size()
print(type_rgb_combinations)
```

```python
# Convert RGBs combinations to ordinal values
df_original["hpf"] = df_original['R'].astype(str) + '-' + df_original['G'].astype(str) + '-' + df_original['B'].astype(str)

print(df_original['hpf'].unique().tolist())

hpfs = df_original['hpf'].unique().tolist()
for i, hpf in enumerate(hpfs):
    df_original.loc[df_original['hpf'] == hpf, 'hpf'] = i

print(df_original['hpf'].unique().tolist())
```
